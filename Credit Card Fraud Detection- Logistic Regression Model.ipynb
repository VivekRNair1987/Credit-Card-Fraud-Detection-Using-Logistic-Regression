##### Dealing with unbalanced data

UnderSampling 

Build a sample dataset containing similar distribution of legit transactions and the fraud transactions

No. of Fraud Transactions = 492

So we would take random 492 transactions from the legit dataset and then join it with fraud transactions so that there would be uniformity in the class feature and the distribution would be even

```python
legit_sample = legit.sample(n=492)
```

```python
## Concatenating two dataframes
new_cc_dataset = pd.concat([legit_sample,fraud], axis=0)
```

```python
## Checking the first 5 rows of the new df

new_cc_dataset.head()
```
